### Obtaining Statistics by Grouping the Variables

```python
grouped_df = df.groupby('groups')['salary'].agg(['mean',    
                                                 'median',  
                                                 'max', 
                                                 'min', 
                                                 'std',
                                                 'var',])   


#NOTE : mean   -> Mean,
#       median -> Median,
#       max    -> Maximum value,
#       min    -> Minimum value,
#       std    -> Standard Value,
#       var    -> Variance,
```

```python
#Displaying the result
grouped_df
```

```python
#Group by species
df_iris.groupby('species').describe()
```

### Displaying some basic statistical details of Species in IRIS Dataset

```python
#Checking unique types of species
df_iris['species'].unique()
```

```python
#Creating seperate dataframes for the species
setosa_data = df_iris[df_iris['species'] == 'setosa']
versicolor_data = df_iris[df_iris['species'] == 'versicolor']
virginica_data = df_iris[df_iris['species'] == 'virginica']
```
