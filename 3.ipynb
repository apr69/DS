### Task - 1

### Loading Dataset

```python
df = load_dataset('titanic')
df
```

### Understanding the Dataset

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].mean(),axis = 0,inplace = True)
df.drop('deck',axis = 1,inplace = True)
cols = ['embarked','embark_town']
for col in cols:
    df[col].fillna(df[col].mode()[0],inplace = True)
```

```python
df.isnull().sum()
```

### Basic Statistical Analysis

```python
df.describe()
```

### Obtaining Statistics by Grouping the Variables

```python
grouped_df = df.groupby('sex')['fare'].agg(['mean',    
                                            'median',  
                                            'max', 
                                            'min', 
                                            'std',
                                            'var',])   


#NOTE : mean   -> Mean,
#       median -> Median,
#       max    -> Maximum value,
#       min    -> Minimum value,
#       std    -> Standard Value,
#       var    -> Variance,
```

```python
#Displaying the result
grouped_df
```
